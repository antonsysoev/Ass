```r
ishigami <- function(x1,x2,x3){
  sin(x1)+7*(sin(x2))^2+0.1*((x3)^4) * sin(x1)
}
```

```r
n <- 3
l <- rep(list((-10:10)), n)

x <- expand.grid(l)
head(x$Var1)
```

```r
head(x)
```

```r
model_rosenbrock <- rosenbrock(x$Var1, x$Var2, x$Var3)
```

```r
mod <- x
mod$y <- model_rosenbrock
```

```r
alpha <- 0
znach_1 <- 0
znach_2 <- 0
znach_3 <- 0


for(i in 2:length(mod$Var1)){
				x_1 = mod$Var1[i-1];
				x_2 = mod$Var2[i-1];
        x_3 = mod$Var3[i-1];
				dx_1 = mod$Var1[i]-mod$Var1[i-1];
				dx_2 = mod$Var2[i]-mod$Var2[i-1];
        dx_3 = mod$Var3[i]-mod$Var3[i-1];


func <- function(a){
      (x_1+dx_1)^2+(x_2+dx_2)^4+(x_1+dx_1)*(x_2+dx_2)+(x_2+dx_2)*(x_3+dx_3)^4 -
(x_1^2+x_2^4+x_1*x_2+x_2*x_3^4) -
(2 * (x_1+a*dx_1) + (x_2+a*dx_2))*dx_1 -
(4 * (x_2+a*dx_2)^3 + (x_1+a*dx_1) + (x_3+a*dx_3)^4)*dx_2 -
((x_2+a*dx_2) * (4 * (x_3+a*dx_3)^3))*dx_3
  }

alpha[i] <-  optimize(func, c(0, 1), tol = 0.1)$`minimum`;
				alpha[i] <- as.data.frame(alpha[i])[1,]

        znach_1[i] <- 2 * (x_1+ alpha[i]*dx_1) + (x_2+ alpha[i]*dx_2)
        znach_2[i] <- 4 * (x_2+ alpha[i]*dx_2)^3 + (x_1+ alpha[i]*dx_1) + (x_3+ alpha[i]*dx_3)^4
        znach_3[i] <- (x_2+ alpha[i]*dx_2) * (4 * (x_3+ alpha[i]*dx_3)^3)
}
```

```r
res_1=data.frame(X_1 = mean(znach_1[2:length(znach_1)]), X_2 = mean(znach_2[2:length(znach_2)]), X_3 = mean(znach_3[2:length(znach_3)]))
```

```r
dev_ros_x1 <- 0
dev_ros_x2 <- 0
dev_ros_x3 <- 0

for(i in 1:length(x$Var1)){
dev_ros_x1[i] <- derivative(f = rosenbrock, var = c(x1=x$Var1[i], x2=x$Var2[i], x3=x$Var3[i]), order = c(1,0,0))
dev_ros_x2[i] <- derivative(f = rosenbrock, var = c(x1=x$Var1[i], x2=x$Var2[i], x3=x$Var3[i]), order = c(0,1,0))
dev_ros_x3[i] <- derivative(f = rosenbrock, var = c(x1=x$Var1[i], x2=x$Var2[i], x3=x$Var3[i]), order = c(0,0,1))
}
```

```r
dev_ros_x1
```

```r
dev_x1 <- cbind(x,dev_ros_x1)
dev_x2 <- cbind(x,dev_ros_x2)
dev_x3 <- cbind(x,dev_ros_x3)
colnames(dev_x1) <- c('x1','x2','x3', 'proiz')
colnames(dev_x2) <- c('x1','x2','x3', 'proiz')
colnames(dev_x3) <- c('x1','x2','x3', 'proiz')
```

```r
lm.1 <- lm(proiz ~ (x1*x2*x3), dev_x1)
lm.2 <- lm(proiz ~ (x1*x2*x3), dev_x2)
lm.3 <- lm(proiz ~ (x1*x2*x3), dev_x3)
```
