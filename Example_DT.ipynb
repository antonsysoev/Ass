```r
install.packages("tree")
install.packages("mlbench")
library(tree)
library(mlbench)

data('BostonHousing')
mydata <- BostonHousing
```

```r
install.packages("rpart")
library(rpart)
install.packages("rpart.plot")
library(rpart.plot)
```

```r
set.seed(1234)
ind <- sample(2, nrow(mydata), replace = T, prob = c(0.5, 0.5))
train <- mydata[ind == 1,]
test <- mydata[ind == 2,]

tree <- rpart(medv ~., data = train)
rpart.plot(tree)
```
